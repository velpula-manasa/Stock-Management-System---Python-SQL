```python
import sqlite3
from tkinter import messagebox,Tk

conn=sqlite3.connect("stock.db")
cursor=conn.cursor()
## create table 
cursor.execute("""
  create table IF NOT EXISTS Stock (
  ProductID INTEGER PRIMARY KEY AUTOINCREMENT,
  ProductName VARCHAR(100),
  Category VARCHAR(50),
  QuantityInStock INT,
  UnitPrice DECIMAL(10,2),
  MinStock INT
)""")

## Insert values in table 
cursor.execute("""
insert into Stock(ProductName,Category,QuantityInStock,UnitPrice,MinStock) values(?,?,?,?,?)
""",('Rice','grocery',10,2000.00,3))

conn.commit()


## fetching records
"""cursor.execute("select * from stock")
rows=cursor.fetchall()

print("productname | Category | QuantityInStock | UnitPrice | MinStock")
for row in rows:
    print(row)
conn.commit()"""

## Adding products
def add_product():
    product_name=input("Enter any productname:")
    category=input("Enter any category:")
    quantity=int(input("Enter any quantity:"))
    unit_price=float(input("Enter any unitprice:"))
    min_stock=int(input("Enter minimum stock:"))

    cursor.execute("""
     insert into Stock(ProductName,Category,QuantityInStock,UnitPrice,MinStock) values(?,?,?,?,?)
     """,(product_name,category,quantity,unit_price,min_stock))
    conn.commit()
    print("Product added successfully")


## Viewing products
def view_product():
   cursor.execute("select * from stock")
   rows=cursor.fetchall()

   print("productname | Category | QuantityInStock | UnitPrice | MinStock")
   print("----------------------------------------------------------------------")
   for row in rows:
    print(row)

## Updating products
def update_product():
   product_name1=input("Enter any productname:")
   quantity=int(input("Enter any quantity to add:"))
   unit_price=float(input("Enter any unit_price:"))

   cursor.execute("""
      update Stock set QuantityInStock=QuantityInStock+?,
      UnitPrice=?
      where productname=?
      """,(quantity,unit_price,product_name1))
   if cursor.rowcount>0:
       conn.commit()
       print("Product updated Successfully")
   else:
      print("Product not available")

## update Minstock
def update_min_stock():
    qty_used=int(input("Enter any quantity used:"))
    product_name2=input("Enter any product_name:")

    cursor.execute("select QuantityInStock,MinStock from Stock where ProductName=?",(product_name2,))
    record=cursor.fetchone()

    if record is None:
        print("No record found")
        return
    availability,min_stock1=record   ## unpacking

    if qty_used>availability:
        messagebox.showerror("Error"," Not enough stock")
        return
    

    new_qty=availability-qty_used
    cursor.execute("Update Stock set QuantityInStock=? where ProductName=?",(new_qty,product_name2))
    conn.commit()
    
    if new_qty<min_stock1:
        messagebox.showwarning("Low Stock Alert","Please check your Stock")
    messagebox.showinfo("Success",f"Successfully used {qty_used} units of {product_name2} and remaining {new_qty}")


def Management():
  while True:
    print(""" Add your choice:\n
              1.Add product
              2.view product
              3.update_product
              4.update_min_stock
              5.Exit""")
    choice=int(input("Enter your choice:"))
    if choice==1:
        add_product()
    elif choice==2:
        view_product()
    elif choice==3:
        update_product()
    elif choice==4:
        update_min_stock()
    else:
       print("Exiting..")
       break
Management()
conn.close()
               
```
